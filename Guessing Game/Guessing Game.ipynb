```python
import random
num=random.randint(1,100)

print("WELCOME TO GUESS ME!")
print("I'll think of a number between 1 and 100")
print("If your guess is more than 10 numbers away from my number, I'll tell you you're COLD")
print("If your guess is within 10 numbers of my number, I'll tell you you're WARM")
print("If your guess is farther than your most recent guess, I'll say you're getting COLDER")
print("If your guess is closer than your most recent guess, I'll say you're getting WARMER")
print("LET'S GUESS AND PLAY!")

YourGuess=[0]

while True:
    print("I'm thinking of a number between 1 and 100.")
    guess=int(input("What's your guess?"))
    
    if guess<1 or guess>100:
        print("OUT OF BOUNDS!")
        print("Try again.")
        continue
    
    if guess==num:
        print ("CONGRATULATIONS!YOU GUESSED IT RIGHT!")
        break
    
    YourGuess.append(guess)
    
    if YourGuess[-2]:
        if abs(num-guess)>abs(num-YourGuess[-2]):
            print("COLDER!")
        else:
            print("WARMER!")
    
    else:
        if abs(num-guess)>10:
            print('COLD!')
        else:
            print('WARM!')
```
